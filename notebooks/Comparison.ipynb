```python
#Grafik Perbadingan MSE dan RMSE
import matplotlib.pyplot as plt
import numpy as np

# Data model berdasarkan batch size dan epoch yang diperbarui
models = ['BS16-E100', 'BS4-E50', 'BS32-E100', 'BS64-E100', 'BS128-E100']
mse_standard = [0.006984, 0.006990, 0.006999, 0.007006, 0.007008]
mse_attention = [0.006849, 0.006894, 0.006915, 0.006956, 0.006985]
rmse_standard = [0.083569, 0.083604, 0.083660, 0.083702, 0.083716]
rmse_attention = [0.082757, 0.083031, 0.083155, 0.083403, 0.083575]

x = np.arange(len(models))
width = 0.35

# Plot MSE
fig, ax = plt.subplots(figsize=(8, 4))
bar1 = ax.bar(x - width/2, mse_standard, width, label='LSTM Standar', color='mediumseagreen')
bar2 = ax.bar(x + width/2, mse_attention, width, label='LSTM with Attention Mechanism', color='deepskyblue')

ax.set_ylabel('MSE')
ax.set_title('MSE Comparison')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)
ax.set_ylim(0.0065, 0.0072)

plt.tight_layout()
plt.show()

# Plot RMSE
fig, ax = plt.subplots(figsize=(8, 4))
bar3 = ax.bar(x - width/2, rmse_standard, width, label='LSTM Standar', color='goldenrod')
bar4 = ax.bar(x + width/2, rmse_attention, width, label='LSTM with Attention Mechanism', color='crimson')

ax.set_ylabel('RMSE')
ax.set_title('RMSE Comparison')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)
ax.set_ylim(0.0825, 0.084)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Ags', 'Sep', 'Okt', 'Nov', 'Des']
hujan_aktual = [55.00, 55.00, 55.00, 55.00, 55.00, 55.08896552, 55.00, 55.00, 55.00, 58.1137931, 55.00, 55.00]
prediksi_lstm = [55.56839357, 55.6400568, 55.43080549, 55.55709634, 54.88772956, 55.8373892,
                 56.17450976, 57.12571673, 56.96777524, 58.13772038, 57.04052102, 57.615691]
prediksi_attention = [55.62893803, 55.59224548, 55.58249615, 55.57365829, 55.56720417, 55.70716975, 55.88693663, 58.83585843, 57.01629394, 58.96442491, 56.91476311, 56.47451745]

# Konversi ke array
x = np.arange(len(bulan))
width = 0.25

# Plot
fig, ax = plt.subplots(figsize=(12, 6))
bar1 = ax.bar(x - width, hujan_aktual, width, label='Hujan Aktual', color='forestgreen')
bar2 = ax.bar(x, prediksi_lstm, width, label='Prediksi LSTM Standar', color='royalblue')
bar3 = ax.bar(x + width, prediksi_attention, width, label='Prediksi LSTM dengan Mekanisme Attention', color='darkorange')

# Tambahkan label di atas batang
for i in range(len(bulan)):
    ax.annotate(f"{hujan_aktual[i]:.1f}", xy=(x[i] - width, hujan_aktual[i]), xytext=(0, 2),
                textcoords="offset points", ha='center', va='bottom', fontsize=8, color='green')
    ax.annotate(f"{prediksi_lstm[i]:.1f}", xy=(x[i], prediksi_lstm[i]), xytext=(0, 2),
                textcoords="offset points", ha='center', va='bottom', fontsize=8, color='blue')
    ax.annotate(f"{prediksi_attention[i]:.1f}", xy=(x[i] + width, prediksi_attention[i]), xytext=(0, 2),
                textcoords="offset points", ha='center', va='bottom', fontsize=8, color='orange')

# Pengaturan sumbu dan label
ax.set_xticks(x)
ax.set_xticklabels(bulan)
ax.set_ylabel("Rainfall (mm)")

# Legend dekat dan rata
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08), ncol=3, frameon=False, fontsize=10)

# Tambahan: grid dan layout
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.subplots_adjust(bottom=0.2)  # atur space bawah supaya legend tidak terpotong
plt.tight_layout()
plt.show()
```
